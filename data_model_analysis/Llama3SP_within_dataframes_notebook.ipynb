```python
import pandas as pd
```

```python
og_within_df = pd.read_csv('og-within-project.csv')
og_within_df
```

```python
df_within_mae = pd.read_csv('within_results.csv')
df_within_mdae = pd.read_csv('Llama3_within_results_mdae.csv')
df_within_sa = pd.read_csv('Llama3_within_results_sa.csv')

df_within_mae
```

```python
df_within_mae = df_within_mae[['Project', 'Llama3SP']]
df_within_mae = df_within_mae.rename(columns={'Llama3SP': 'MAE'})
df_within_mdae = df_within_mdae[['Project', 'Llama3.2']]
df_within_mdae = df_within_mdae.rename(columns={'Llama3.2': 'MdAE'})
df_within_sa = df_within_sa[['Project', 'Llama3.2']]
df_within_sa = df_within_sa.rename(columns={'Llama3.2': 'SA'})

df_within = pd.merge(df_within_mae, df_within_mdae, on='Project')
df_within = pd.merge(df_within, df_within_sa, on='Project')

df_within['Method'] = 'Llama3SP'

df_within
```

```python
dictionary = {
    'appceleratorstudio': 'Appcelerator Studio',
    'aptanastudio': 'Aptana Studio',
    'bamboo' : 'Bamboo',
    'clover' : 'Clover', 
    'datamanagement' : 'Data Management',
    'duracloud' : 'Duracloud',
    'jirasoftware' : 'Jira Software',
    'mesos' : 'Mesos',
    'moodle' : 'Moodle',
    'mule' : 'Mule',
    'mulestudio' : 'Mulesoft',
    'springxd' : 'Spring XD',
    'talenddataquality' : 'Talend Data Quality',
    'talendesb' : 'Talend ESB',
    'titanium' : 'Titanium',
    'usergrid' : 'Usergrid',
}
```

```python
# Add the project name
df_within['Project'] = df_within['Project'].map(dictionary)

df_within
```

```python
final_df = pd.concat([og_within_df, df_within])
final_df = final_df[['Project', 'Method', 'MAE', 'MdAE', 'SA']]
final_df = final_df.sort_values(by=['Project', 'Method'])

final_df
```

```python
# Function to hide repeated values and highlight the minimum MAE, MdAE, and SA
def hide_repeated_and_highlight_min_metrics(df):
    # Sort the DataFrame to ensure 'Llama3SP' is first within each project
    df = df.sort_values(by=['Project', 'Method'], key=lambda col: col.map(lambda x: (x != 'Llama3SP', x)))
    
    df_copy = df.copy()  # Create a copy of the original DataFrame
    styles = pd.DataFrame('', index=df.index, columns=df.columns)

    for project in df['Project'].unique():
        mask = (df['Project'] == project)
        if mask.any():
            min_mae = df[mask]['MAE'].min()
            min_mdae = df[mask]['MdAE'].min()
            min_sa = df[mask]['SA'].min()
            styles.loc[mask & (df['MAE'] == min_mae), 'MAE'] = 'font-weight: bold'
            styles.loc[mask & (df['MdAE'] == min_mdae), 'MdAE'] = 'font-weight: bold'
            styles.loc[mask & (df['SA'] == min_sa), 'SA'] = 'font-weight: bold'
    
    mask = (df_copy[['Project']] != df_copy[['Project']].shift()).any(axis=1)
    df_copy.loc[~mask, 'Project'] = ''
    
    return df_copy, styles

# Apply the function to hide repeated values and highlight the minimum metrics to DataFrame
within_df_copy, within_styles = hide_repeated_and_highlight_min_metrics(final_df)

# Reset the index to ensure unique indices
within_df_copy = within_df_copy.reset_index(drop=True)
within_styles = within_styles.reset_index(drop=True)

# Apply the styles to each DataFrame
styled_within_df = within_df_copy.style.apply(lambda x: within_styles, axis=None)

# Round the values of MAE, MdAE, SA, and 'Llama3SP vs.' to 2 decimal places in each DataFrame
styled_within_df = styled_within_df.format({'MAE': '{:.2f}', 'MdAE': '{:.2f}', 'SA': '{:.2f}'})

styled_within_df
```

```python
styled_within_df.to_excel('llama3sp_final_within_results.xlsx', engine='openpyxl', index=False)
```
