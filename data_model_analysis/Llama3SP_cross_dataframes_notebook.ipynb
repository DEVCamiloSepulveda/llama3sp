```python
import pandas as pd
```

```python
og_cross_df = pd.read_csv('og-cross-repository.csv')
og_within_df = pd.read_csv('og-within-repository.csv')

additional_transformed_data = []

og_cross_df = og_cross_df.drop(columns=['GPT2SP_vs'])
og_cross_df['Scenario'] = 'cross'
og_cross_df = og_cross_df[['Scenario', 'Source', 'Target', 'Method', 'MAE', 'MdAE', 'SA']]
og_within_df = og_within_df.drop(columns=['GPT2SP_vs'])
og_within_df['Scenario'] = 'within'
og_within_df = og_within_df[['Scenario', 'Source', 'Target', 'Method', 'MAE', 'MdAE', 'SA']]

og_df = pd.concat([og_cross_df, og_within_df])

og_df
```

```python
df_cross = pd.read_csv('cross_results.csv')
df_cross
```

```python
df_cross = df_cross[['Train', 'Test', 'Llama3SP']]
df_cross
```

```python
df_llama3_cross_mdae = pd.read_csv('Llama3_cross_results_mdae.csv')
df_llama3_cross_mdae = df_llama3_cross_mdae.rename(columns={'Llama3.2': 'MdAE'})
df_llama3_cross_mdae = df_llama3_cross_mdae[['Train', 'Test', 'MdAE']]

df_llama3_cross_mdae
```

```python
df_llama3_cross_sa = pd.read_csv('Llama3_cross_results_sa.csv')
df_llama3_cross_sa = df_llama3_cross_sa.rename(columns={'Llama3.2': 'SA'})
df_llama3_cross_sa = df_llama3_cross_sa[['Train', 'Test', 'SA']]

df_llama3_cross_sa
```

```python
df_cross = pd.merge(df_cross, df_llama3_cross_mdae, how='left', left_on=['Train', 'Test'], right_on=['Train', 'Test'])
df_cross = pd.merge(df_cross, df_llama3_cross_sa, how='left', left_on=['Train', 'Test'], right_on=['Train', 'Test'])

df_cross
```

```python
df = df_cross.copy()

# Create a list to store the transformed data
transformed_data = []

for _, row in df.iterrows():
    train, test = row['Train'], row['Test']

    row['Scenario'] = og_df.loc[(og_df['Source'] == train) & (og_df['Target'] == test), 'Scenario'].values[0]
    row['Source'] = train
    row['Target'] = test
    row['Method'] = 'Llama3SP'
    row['MAE'] = row['Llama3SP']
    row['MdAE'] = row['MdAE']
    row['SA'] = row['SA']
    row['Llama3SP vs.'] = 0.0
    transformed_data.append(row)

# Create the transformed DataFrame
transformed_df = pd.DataFrame(transformed_data)

# Reorder and select columns
cross_project_df = transformed_df[[
    'Scenario', 'Source', 'Target', 'Method', 'MAE', 'MdAE', 'SA', 'Llama3SP vs.'
]]

# Print the DataFrame
cross_project_df
```

```python
final_df = cross_project_df.copy()

# Iterate through the rows of og_df
for _, row in og_df.iterrows():
    scenario, source, target, method, mae, mdae, sa = row['Scenario'], row['Source'], row['Target'], row['Method'], row['MAE'], row['MdAE'], row['SA']

    # Comparison with Llama3SP
    llama3sp_mae = df.loc[(df['Train'] == source) & (df['Test'] == target), 'Llama3SP'].values[0]
    llama3sp_mdae = df.loc[(df['Train'] == source) & (df['Test'] == target), 'MdAE'].values[0]
    llama3sp_sa = df.loc[(df['Train'] == source) & (df['Test'] == target), 'SA'].values[0]

    mae_diff = ((mae - llama3sp_mae) / llama3sp_mae) * 100
    mdae_diff = ((mdae - llama3sp_mdae) / llama3sp_mdae) * 100
    sa_diff = sa - llama3sp_sa

    # Append the new row to final_df
    final_df = final_df.append({
        'Scenario': scenario,
        'Source': source,
        'Target': target,
        'Method': method,
        'MAE': mae,
        'MdAE': mdae,
        'SA': sa,
        'Llama3SP vs.': round(mae_diff, 2)
    }, ignore_index=True)

# Order the data by Scenario, Source, and Target (Scenario DESC, Source ASC, Target ASC)
final_df = final_df.sort_values(by=['Scenario', 'Source', 'Target'], ascending=[False, True, True]).reset_index(drop=True)

# Print the DataFrame
final_df
```

```python
# Function to hide repeated values and highlight the minimum MAE, MdAE, and SA
def hide_repeated_and_highlight_min_metrics(df):
    df_copy = df.copy()  # Create a copy of the original DataFrame
    styles = pd.DataFrame('', index=df.index, columns=df.columns)
    
    for scenario in df['Scenario'].unique():
        for source in df['Source'].unique():
            for target in df['Target'].unique():
                mask = (df['Scenario'] == scenario) & (df['Source'] == source) & (df['Target'] == target)
                if mask.any():
                    min_mae = df[mask]['MAE'].min()
                    min_mdae = df[mask]['MdAE'].min()
                    min_sa = df[mask]['SA'].min()
                    styles.loc[mask & (df['MAE'] == min_mae), 'MAE'] = 'font-weight: bold'
                    styles.loc[mask & (df['MdAE'] == min_mdae), 'MdAE'] = 'font-weight: bold'
                    styles.loc[mask & (df['SA'] == min_sa), 'SA'] = 'font-weight: bold'
    
    mask = (df_copy[['Scenario', 'Source', 'Target']] != df_copy[['Scenario', 'Source', 'Target']].shift()).any(axis=1)
    df_copy.loc[~mask, ['Scenario', 'Source', 'Target']] = ''
    
    return df_copy, styles

# Separate the data into two different DataFrames
within_df = final_df[final_df['Scenario'] == 'within']
cross_df = final_df[final_df['Scenario'] == 'cross']

# Apply the function to hide repeated values and highlight the minimum metrics to each DataFrame
within_df_copy, within_styles = hide_repeated_and_highlight_min_metrics(within_df)
cross_df_copy, cross_styles = hide_repeated_and_highlight_min_metrics(cross_df)

# Remove the 'Scenario' column from each DataFrame
within_df_copy = within_df_copy.drop(columns=['Scenario'])
cross_df_copy = cross_df_copy.drop(columns=['Scenario'])

# Adjust the style DataFrames to match the DataFrames without the 'Scenario' column
within_styles = within_styles.drop(columns=['Scenario'])
cross_styles = cross_styles.drop(columns=['Scenario'])

# Apply the styles to each DataFrame
styled_within_df = within_df_copy.style.apply(lambda x: within_styles, axis=None)
styled_cross_df = cross_df_copy.style.apply(lambda x: cross_styles, axis=None)

# Round the values of MAE, MdAE, SA, and 'Llama3SP vs.' to 2 decimal places in each DataFrame
styled_within_df = styled_within_df.format({'MAE': '{:.2f}', 'MdAE': '{:.2f}', 'SA': '{:.2f}', 'Llama3SP vs.': '{:.2f}'})
styled_cross_df = styled_cross_df.format({'MAE': '{:.2f}', 'MdAE': '{:.2f}', 'SA': '{:.2f}', 'Llama3SP vs.': '{:.2f}'})

styled_within_df, styled_cross_df
```
